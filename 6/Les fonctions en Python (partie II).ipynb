<h2 style="color:blue">6. La fonction Lambda </h2>

Python permet de créer des fonctions anonymes limitées à une seule instruction plus rapidement qu'avec l'instruction **def**.

On utilise pour cela l'instruction `lambda` de la façon suivante :

```python
# Définissons notre première fonction lambda f
f = lambda x: x * x

#Utilisons f
f(5)
```

<h3 style="color:#891bbf">Syntaxe</h3>

Nous définissons la fonction lambda avec Python selon la syntaxe suivante :

`lambda paramètres : expression ` 

La fonction lambda retourne la valeur de l'expression évaluée comme n'importe quelle fonction Python classique (en l'absence de "return").

Les fonctions **lambda** sont très utilisées en mathématiques. Ce sont des fonctions anonymes (elle n'ont pas de nom) stockées dans des **variables**.

Vous pouvez utiliser plusieurs paramètres :

```python
#Définissons la fonction lambda s
s = lambda x, y: x + y

#Utilisons s maintenant
s(4,5)
```

Dans cet exemple, nous créons une fonction lambda qui prend deux arguments `x` et `y` et renvoie leur somme.

Nous pouvons même renvoyer leur somme en une seule ligne comme suit :

```python
(lambda x, y: x + y)(4, 5)
```

<h2 style="color:blue">6. Les fonctions récursives </h2>

Une fonction récursive est une fonction qui s'appelle elle-même pour résoudre un problème. C'est une idée puissante pour résoudre des problèmes complexes.

Pour qu’une fonction récursive fonctionne correctement, deux éléments sont essentiels :

- **Le cas de base** : Une condition qui met fin à la récursion.
- **L’appel récursif** : Lorsque la fonction s'appelle elle-même en progressant vers le cas de base.

Cela vous paraît peut-être compliqué au début, mais nous allons essayer de créer notre première fonction récursive pour comprendre mieux ce nouveau concept. 

On souhaite créer une fonction **countDown(nbDepart)** qui affiche le compte à rebours à partir d'un nombre passé en argument :

```python
def countDown(nbDepart):
    print(nbDepart)

countDown(5)
```

Nous remarquons que la fonction n'affiche que le numéro de départ.

Pour afficher 5, 4, 3, 2 et 1 dans l’ordre, nous devons :

- Tout d’abord, appeler countDown(5) pour afficher 5.
- Puis, appeler countDown(4) pour afficher 4.
- Appeler countDown(3) pour afficher 3.
- Appeler countDown(2) pour afficher 2.
- Enfin, appeler countDown(1) pour afficher 1.

Pour ce faire, à l’intérieur de la fonction countDown(), nous devons définir une logique pour appeler la fonction countDown() avec les arguments 4, 3, 2 et 1. Donc, nous devons transformer la fonction countDown() en une fonction récursive.

Nous redéfinissons la fonction countDown() et nous l’appellons en passant le nombre 5 comme suit:

```python
def countDown(nbDepart):
    """ fonction qui affiche le compte à rebours à partir de nbDepart

    param nbDepart : entier  
    """

    print(nbDepart)
    countDown(nbDepart-1)

countDown(5)
```

Vous devez normalement avoir l'erreur suivante : `RecursionError: maximum recursion depth exceeded `

La raison est que **countDown()** s’appelle indéfiniment jusqu’à ce que le système l’arrête. En effet, Python ne permet pas de dépasser les 3000 appels.

Nous pouvons vérifier cette limite à l’aide de ce paramètre système **sys.getrecursionlimit()** :


```python
import sys 

print(sys.getrecursionlimit())
```

Revenons à notre fonction, nous devons l'interrompre une fois qu'elle atteigne zéro. Par conséquent, nous devons rajouter une condition d'arrêt comme celle-ci :

```python
def countDown(nbDepart):
    """ fonction qui affiche le compte à rebours à partir de nbDepart

    param nbDepart : entier  
    """

    print(nbDepart)
    nbSuivant = nbDepart - 1
    if nbSuivant > 0:
          countDown(nbSuivant)

    

countDown(5)
```

Nous réussissons finalement à créer notre première fonction récursive. Comme vous avez pu le voir, la fonction countDown() s'appelle uniquement lorsque le nombre suivant est supérieur à zéro (c'est pourquoi qu'on l'appelle "condition d'arrêt"). Une fois à zéro, elle s'arrête de s'appeler.
